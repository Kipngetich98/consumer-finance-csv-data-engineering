```python
import pandas as pd

# Read the data and the first 100 rows

# df = pd.read_csv('/home/vincent/Downloads/SCFP2019.csv', nrows=100)
df = pd.read_csv("/app/SCFP2019.csv", nrows=100)
# nrows -> first n amount of rows you want to read
df.head()
```

put the data to postgres

```python
#generate a schema of the dataset, which is an instruction to create a table that specifies what kind of columns you want to have there and their types
pd.io.sql.get_schema(df, name="consumer_data")

#The get_schema function converts your Pandas "Dataframe" to a DDL which is used for specifying schemas. This function outputs a CREATE TABLE sql statement with table name being "consumer_data"
print(pd.io.sql.get_schema(df, name="consumer_data"))
```

# Putting the schema into postgresql

```python
from sqlalchemy import create_engine

# specify the database you want to use based on the docker run command we had
# postgresql://username:password@localhost:port/dbname
engine = create_engine('postgresql://postgres:1234@localhost:5435/consumer_data')

# create the connection to the database engine to see if everything is working properly
engine.connect()

# pass the engine variable to get_schema function
# Pandas will execute the schema SQL statement using the engine connection we have defined
pd.io.sql.get_schema(df, name="consumer_data", con=engine)
```
